```python
import pandas as pd
```

```python
response = rq.get('https://api.nasa.gov/neo/rest/v1/feed?start_date=2025-02-23&end_date=2025-03-01&api_key=mgo2O93jxuor26UNUcOIhterpZ7QNfYhCKdIgNBR')
data1 = response.json()
print(f'the request of {data1} was successful')
```

```python
asteroids = []
for dat in data1["near_earth_objects"]:  # Loop through dat
    for asteroid in data1["near_earth_objects"][dat]:  # Loop through asteroids on each dat
        asteroids.append({
            "Asteroid ID": asteroid["id"],
            "Asteroid Name": asteroid["name"],
            "Minimal Estimated Diameter (km)": asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
            "Absolute Magnitude": asteroid["absolute_magnitude_h"],
            "Relative Velocity (km/s)": asteroid["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"]
        })
```

```python
df = pd.DataFrame(asteroids)
df.head(20)
```
